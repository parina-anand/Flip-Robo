Q2.
Scrape the details team Indiaâ€™s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the homepage of BCCI
url = "https://www.bcci.tv/"
driver.get(url)
```

```python
international_button = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]')       # Locating international_button by xpath
international_button.click()  # Clicking the button to view drop-down
```

```python
fixers = driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]')
url = fixers.get_attribute("href")
driver.get(url) # navigating to inetrnational fixtures page
```

```python
match_title=[]
match_series=[]
place=[]
date=[]
time=[]

loc = driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]//span')
for i in loc:
    place.append(i.text)
    
ser = driver.find_elements_by_xpath('//div[@class="fixture__description u-unskewed-text"]//p/strong')
for k in ser:
    match_title.append(k.text)

t = driver.find_elements_by_xpath('//span[@class="fixture__time"]')
for j in t:
    time.append(j.text)
    
s = driver.find_element_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]')
match_series.append(s.text)
 
p = driver.find_elements_by_xpath('//div[@class="fixture__full-date"]/span')
p2 =  driver.find_elements_by_xpath('//div[@class="fixture__full-date"]/div')
d1=[]
d2=[]
for j in p:
    d1.append(j.text)


for k in p2:
    d2.append(k.text.split('\n')[0])


for i in range(len(d1)):
    date.append(d1[i]+' '+d2[i])



# t = driver.find_elements_by_xpath('//div[@class="fixture__full-date"]')
# for j in t:
#     print(j.text)
```

```python
import pandas as pd
 # creating the dataframe from the scraped data 

print(len(match_series),len(match_title),len(place),len(date),len(time))
df=pd.DataFrame({"Match Title":match_title,"Match Series":match_series,"Place":place,"date":date,"time":time})
df
```

```python
# Obtain the number of rows in body
rows = 1+len(driver.find_elements_by_xpath('/html/body/div[3]/div[2]/div[5]/div[1]/div/table/tbody/tr'))
rows
```

```python
# Obtain the number of columns in body
cols = 1+len(driver.find_elements_by_xpath('/html/body/div[3]/div[2]/div[5]/div[1]/div/table/tbody/tr[1]/td'))
cols
```

```python
# Empty lists
gdp = {}
gdp['Rank'] = []
gdp['State'] = []
gdp['GSDP19'] = []
gdp['GSDP18'] = []
gdp['Share'] = []
gdp['GDP'] = []
```

```python
try:
    # finding rank
    dt = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[1]')
    for d in dt:
        gdp['Rank'].append(d.text)
except NoSuchElementException:
    gdp['Rank'].append('-')
    
try:
    # finding state
    dt = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[2]')
    for d in dt:
        gdp['State'].append(d.text)
except NoSuchElementException:
    gdp['State'].append('-')
    
    
try:
    # finding GSDP(19-20)
    dt = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[3]')
    for d in dt:
        gdp['GSDP19'].append(d.text)
except NoSuchElementException:
    gdp['GSDP19'].append('-')
    
    
try:
    # finding GSDP(18-19)
    dt = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[4]')
    for d in dt:
        gdp['GSDP18'].append(d.text)
except NoSuchElementException:
    gdp['GSDP18'].append('-')
    
try:
    # finding share
    dt = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[5]')
    for d in dt:
        gdp['Share'].append(d.text)
except NoSuchElementException:
    gdp['Share'].append('-')
    
try:
    # finding gdp
    dt = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[6]')
    for d in dt:
        gdp['GDP'].append(d.text)
except NoSuchElementException:
    gdp['GDP'].append('-')
```

```python
print(len(gdp['Rank']),len(gdp['State']),len(gdp['GSDP18']),len(gdp['GSDP19']),len(gdp['GDP']),len(gdp['Share']))
data = pd.DataFrame(gdp)
data
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the homepage
url = " https://www.imdb.com/list/ls095964455/"
driver.get(url)
```

```python
tv={}
tv['Name']=[]
tv['Year span']=[]
tv['Genre']=[]
tv['Run time']=[]
tv['Ratings']=[]
tv['Votes']=[]
```

```python
try:
    
    dt = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//a')
    for d in dt:
        tv['Name'].append(d.text)
except NoSuchElementException:
    tv['Name'].append('-')
    
try:
   
    dt = driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')
    for d in dt:
        tv['Year span'].append(d.text)
except NoSuchElementException:
    tv['Year span'].append('-')
    
    
try:
   
    dt = driver.find_elements_by_xpath('//span[@class="genre"]')
    for d in dt:
        tv['Genre'].append(d.text)
except NoSuchElementException:
    tv['Genre'].append('-')
    
    
try:
    
    dt = driver.find_elements_by_xpath('//span[@class="runtime"]')
    for d in dt:
        tv['Run time'].append(d.text)
except NoSuchElementException:
    tv['Run time'].append('-')
    
try:
   
    dt = driver.find_elements_by_xpath('//span[@class="ipl-rating-star__rating"]')
    for d in dt:
        tv['Ratings'].append(d.text)
except NoSuchElementException:
    tv['Ratings'].append('-')
    
try:
    
    dt = driver.find_elements_by_xpath('//p[@class="text-muted text-small"]/span')
    for d in dt:
        tv['Votes'].append(d.text)
except NoSuchElementException:
    tv['Votes'].append('-')
```
