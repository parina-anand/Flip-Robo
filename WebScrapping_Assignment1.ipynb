### Q1 Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def wikiHeaders(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    soup.prettify()
    
    h1_data = []
    h2_data = []
    h3_data = []
    h4_data = []
    h5_data = []
    h6_data = []
    
    # h1
    h1 = soup.find_all(['h1'])   
    for i in h1:
        h1_data.append(i.text.replace('\n',''))
        
    # h2
    h2 = soup.find_all(['h2'])
    for i in h2:
        h2_data.append(i.text.replace('\n',''))

    # h3
    h3 = soup.find_all(['h3'])
    for i in h3:
        h3_data.append(i.text.replace('\n',''))

    # h4
    h4 = soup.find_all(['h4'])
    for i in h4:
        h4_data.append(i.text.replace('\n',''))

    # h5
    h5 = soup.find_all(['h5'])
    for i in h5:
        h5_data.append(i.text.replace('\n',''))

    # h6
    h6 = soup.find_all(['h6'])
    for i in h6:
        h6_data.append(i.text.replace('\n',''))

    #checking the size of data in each
    print("Lengths of each are : ", len(h1_data) ,len(h2_data),len(h3_data),len(h4_data),len(h5_data),len(h6_data))


    print("Header 1 : ", h1_data)
    print("Header 2 : ", h2_data)
    print("Header 3 : ", h3_data)
    print("Header 4 : ", h4_data)
    print("Header 5 : ", h5_data)
    print("Header 6 : ", h6_data)

wikiHeaders('https://en.wikipedia.org/wiki/Main_Page')
```

### Q 2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
def IMDBTop100Movies(url):
    # Download IMDB's Top 100 movies data
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text)

    movie_list= []
    year_list = []
    ratings_list = []

    # for top 100 movies only as mentioned in ques!
    counter = 0

    for movies in soup.find_all('td', class_= "titleColumn"):
        if(counter<100):
            movie_list.append(movies.text.split('\n')[2].strip())
            year_list.append(movies.text.split('\n')[3].replace('(','').replace(')','').strip())
            counter+=1
        else:
            break
            
    # counter for IMDB
    counter = 0

    for ratings in soup.find_all('td', class_="ratingColumn imdbRating"):
        if(counter<100):
            ratings_list.append(ratings.text.replace('\n',''))
            counter+=1
        else:
            break

    #checking the data size in each column before making the dataFrame
    print(len(movie_list),len(year_list),len(ratings_list))

    data = pd.DataFrame()
    data['Movie Name'] = movie_list
    data['IMDB Rating'] = ratings_list
    data['Year of Release'] = year_list

    display(data)



IMDBTop100Movies('http://www.imdb.com/chart/top')
```

### Q 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def AmazonPhones(url):
    headers= { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
}
            
    response = requests.get(url, headers=headers)
    print(response)
    soup = BeautifulSoup(response.text)
    soup.prettify()
        
    productName = []
    productPrice = []
    avgRating = []
    
    for prod in soup.find_all('span', class_= "a-size-medium a-color-base a-text-normal"):                         
        productName.append(prod.text.split('|')[0].replace('\n',''))
        
    for price in soup.find_all('span', class_="a-price-whole"):
        productPrice.append(price.text)
        
        
    for rating in soup.find_all('span', class_="a-icon-alt"):
        avgRating.append(rating.text)
    
        
    print("Lengths of columns :", len(productName), len(productPrice), len(avgRating))
    
    # Since length might not be the same for all the columns, we are finding the smallest in all and then forming the dataFrame frame
    
    
    smallest = len(productName)
     
    if (len(productPrice) < len(productName)) and (len(productPrice) < len(avgRating)):
        smallest = len(productPrice)
    elif (len(avgRating) < len(productName)):
        smallest = len(avgRating)


    data = pd.DataFrame()
    data['Product Name'] = productName[0:smallest]
    data['Product Price'] = productPrice[0:smallest]
    data['Average Ratings'] = avgRating[0:smallest]

    display(data)
    
        
AmazonPhones('https://www.amazon.in/s?k=mobiles+under+20000&ref=nb_sb_noss_2')
```

### Q 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
def Temp(url):
    response = requests.get(url)
    print(response)
    soup = BeautifulSoup(response.text)
    soup.prettify()
    
    period=[]
    temp=[]
    shortDesp=[]    
    
    for s in soup.find_all('div', class_="tombstone-container"):
        period.append(s.text.split('\n')[1])
        temp.append(s.text.split('\n')[2].split(':')[1])
        shortDesp.append(s.text.split('\n')[2].split(':')[0])
        
           
    # checking lenghth of all columns before making the DataFrame
    print("Length of all columns",len(period), len(temp),len(shortDesp))
    
    data = pd.DataFrame()
    
    data['Period']= period
    data['Temp']= temp
    data['Short Description']=shortDesp
    
    
    display(data)
    
Temp('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YNXs0OgzZPY')
```
